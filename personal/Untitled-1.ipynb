```python
import pandas as pd
bt_pokemon = pd.read_csv('C:\\Users\\barre\\Downloads\\Untitled spreadsheet - Sheet1.csv', header = None)
bt_pokemon.rename(columns = {0 : 'Lead', 1 : 'Other_1', 2 : 'Other_2'}, inplace = True)
bt_pokemon['Lead'] = bt_pokemon['Lead'].map(lambda x: x.split("-",1)[0])
bt_pokemon['Other_1'] = bt_pokemon['Other_1'].map(lambda x: x.split("-",1)[0])
bt_pokemon['Other_2'] = bt_pokemon['Other_2'].map(lambda x: x.split("-",1)[0])
bt_pokemon
lead_mons = pd.DataFrame(bt_pokemon['Lead'].value_counts())
type_chart = pd.read_csv('C:/Users/barre/Documents/GitHub/learning/data/pk_type.csv')
type_chart.rename(columns = {'Unnamed: 0' : 'Attacking_type'})
dict = {0 : 'Normal', 1 : 'Fire', 2 : 'Water', 3 : 'Electric', 4 : 'Grass', 5 : 'Ice', 6 : 'Fighting', 7 : 'Poison', 8 : 'Ground', 9 : 'Flying', 10 : 'Psychic', 11 : 'Bug', 12 : 'Rock', 13 : 'Ghost', 14 : 'Dragon', 15 : 'Dark', 16 : 'Steel', 17 : 'Fairy'}
type_chart = type_chart.rename(dict)
type_chart = type_chart.drop(columns = ['Unnamed: 0'])
pokemon_types = pd.read_csv('C:\\Users\\barre\\python stuff\\data\\194bcff35001e7eb53a2a8b441e8b2c6-92200bc0a673d5ce2110aaad4544ed6c4010f687\\194bcff35001e7eb53a2a8b441e8b2c6-92200bc0a673d5ce2110aaad4544ed6c4010f687\\pokemon.csv')
pokemon_types
```

```python
lead_mons = pd.merge(lead_mons, pokemon_types[['Name', 'Type 1', 'Type 2']], left_index = True, right_on='Name')
lead_mons = lead_mons.rename(columns = {'Lead' : 'Count'})
lead_mons = pd.merge(lead_mons, type_chart, left_on = 'Type 1', right_index = True, how = 'left')
lead_mons = pd.merge(lead_mons, type_chart, left_on = 'Type 2', right_index= True, how = 'left')
lead_mons[['Normal_y', 'Fire_y', 'Water_y', 'Electric_y', 'Grass_y', 'Ice_y', 'Fighting_y',
       'Poison_y', 'Ground_y', 'Flying_y', 'Psychic_y', 'Bug_y', 'Rock_y', 'Ghost_y',
       'Dragon_y', 'Dark_y', 'Steel_y', 'Fairy_y']] = lead_mons[['Normal_y', 'Fire_y', 'Water_y', 'Electric_y', 'Grass_y', 'Ice_y', 'Fighting_y',
       'Poison_y', 'Ground_y', 'Flying_y', 'Psychic_y', 'Bug_y', 'Rock_y', 'Ghost_y',
       'Dragon_y', 'Dark_y', 'Steel_y', 'Fairy_y']].fillna(value=1.0)
```

```python
for type_v in type_chart.columns:
    type_x = type_v + '_x'
    type_y = type_v + '_y'
    type_f = type_v + '_final'
    lead_mons[type_f] = lead_mons[type_x] * lead_mons[type_y]
    lead_mons = lead_mons.drop(columns = [type_x, type_y])
lead_mons

    
```

```python
normal_test = lead_mons.groupby(["Count", "Normal_final"]).count()
normal_test = normal_test.reset_index()
normal_test['Val'] = normal_test['Count'] * normal_test['Name']
scores_dict = {}
for score in [0.00, 0.25, 0.5, 1.0, 2.0, 4.0]:
    scores_dict[score] = normal_test[normal_test['Normal_final'] == score]['Val'].sum()
scores_dict
```

```python
type_effectiveness_df = pd.DataFrame(columns = type_chart.index)
for types in type_chart.index:
    types_f = types + '_final'
    test_df = lead_mons.groupby(['Count', types_f]).count()
    test_df = test_df.reset_index()
    test_df['Val'] = test_df['Count'] * test_df['Name']
    scores_dict = {}
    for score in [0.00, 0.25, 0.5, 1.0, 2.0, 4.0]:
        scores_dict[score] = test_df[test_df[types_f] == score]['Val'].sum()
    type_effectiveness_df[types] = scores_dict.values()
type_effectiveness_df = type_effectiveness_df.rename({0 : 'No Effect', 1 : '4 * Resist', 2 : '2 * Resist', 3 : 'Normal', 4 : '2 * Effective', 5 : '4 * Effective'})
```

```python
def get_super_effective_count(types, lead_mons):
    super_mon_df = pd.DataFrame(columns = lead_mons.columns)
    for i in types:
        type_f = i + '_final'
        super_mon_df = pd.concat([super_mon_df, lead_mons[lead_mons[type_f].isin([2.0, 4.0])]]).drop_duplicates().reset_index(drop = True)
    return super_mon_df['Count'].sum()

def get_normal_or_worse_count(types, lead_mons):
    ineffective_mon_df = lead_mons
    for i in types:
        type_f = i + '_final'
        ineffective_mon_df = ineffective_mon_df[~ineffective_mon_df[type_f].isin([1.0, 2.0, 4.0])]
    return ineffective_mon_df['Count'].sum()

def get_ineffective_count(types, lead_mons):
    ineffective_mon_df = lead_mons
    for i in types:
        type_f = i + '_final'
        ineffective_mon_df = ineffective_mon_df[~ineffective_mon_df[type_f].isin([2.0, 4.0])]
    return ineffective_mon_df['Count'].sum()

def get_super_effective_mons(types, lead_mons):
    super_mon_df = pd.DataFrame(columns = lead_mons.columns)
    for i in types:
        type_f = i + '_final'
        super_mon_df = pd.concat([super_mon_df, lead_mons[lead_mons[type_f].isin([2.0, 4.0])]]).drop_duplicates().reset_index(drop = True)
    return super_mon_df

def get_normal_or_worse_mons(types, lead_mons):
    ineffective_mon_df = lead_mons
    for i in types:
        type_f = i + '_final'
        ineffective_mon_df = ineffective_mon_df[~ineffective_mon_df[type_f].isin([1.0, 2.0, 4.0])]
    return ineffective_mon_df

def get_ineffective_mons(types, lead_mons):
    ineffective_mon_df = lead_mons
    for i in types:
        type_f = i + '_final'
        ineffective_mon_df = ineffective_mon_df[~ineffective_mon_df[type_f].isin([2.0, 4.0])]
    return ineffective_mon_df
```

```python
combinations = []
names = []
def combinationUtil(arr, n, r, index, data, i):
    # Current combination is
    # ready to be printed,
    # print it
    if(index == r):
        vals = []
        name = ''
        for j in range(r):
            print(data[j], end = " ")
            name = name  + data[j] + ' '
            vals.append(data[j])
        combinations.append([vals][0])
        names.append(name[:-1])
        print(" ")
        return

    if(i >= n):
        return
    
    data[index] = arr[i]
    combinationUtil(arr, n, r,index + 1, data, i + 1)

    combinationUtil(arr, n, r, index, data, i + 1)

def printcombination(arr, n, r):
    data = list(range(r))
    combinationUtil(arr, n, r, 0, data, 0)
    return


lat_moveset = ['Psychic', 'Dragon', 'Grass', 'Ice', 'Water', 'Ghost', 'Flying', 'Electric', 'Normal']
all_types = list(type_chart.columns)
printcombination(lat_moveset, len(lat_moveset), 2)
#printcombination(all_types, len(all_types), 4)
```

```python
all_types_df = pd.DataFrame(columns = ['Super Effective Count', 'Normal or Worse Count', 'Ineffective Count', 'Name'])
for types, name in zip(combinations, names):
    print(types)
    x = get_super_effective_count(types, lead_mons)
    y = get_normal_or_worse_count(types, lead_mons)
    z = get_ineffective_count(types, lead_mons)
    row = pd.Series({'Super Effective Count' : x, 'Normal or Worse Count' : y, 'Ineffective Count' : z, 'Name' : name})
    all_types_df = pd.concat([all_types_df, row.to_frame().T], )

```

```python
all_types_df.sort_values(by = ['Super Effective Count', 'Normal or Worse Count', 'Ineffective Count'], ascending= [False, True, True]).to_csv('C:/Users/barre/Documents/GitHub/learning/data/lat_2_types.csv')
```
